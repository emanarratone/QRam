```python
import numpy as np
from qiskit import QuantumRegister, QuantumCircuit
from qiskit.quantum_info import Statevector                               

bus = QuantumRegister(1, name='bus')
ad2 = QuantumRegister(3, name='ad2')
qram = QuantumRegister(12, name='qram')
qc = QuantumCircuit(bus,ad2,qram)

#load address
qc.h(ad2[0])
qc.h(ad2[1])
qc.cx(ad2[1],ad2[2])
qc.x(ad2[1])
qc.swap(ad2[0],qram[0])
qc.x(qram[0])
qc.cswap(qram[0],ad2[1],qram[1])
qc.cswap(qram[0],ad2[2],qram[2])
qc.x(qram[0])
qc.cswap(qram[0],ad2[1],qram[3])
qc.cswap(qram[0],ad2[2],qram[4])

qc.barrier()
qc.cx(qram[1],qram[5])
qc.cx(qram[2],qram[7])
qc.cx(qram[3],qram[9])
qc.cx(qram[4],qram[11])
qc.barrier()

qc.x(qram[0])
qc.cswap(qram[0],ad2[1],qram[1])
qc.cswap(qram[0],ad2[2],qram[2])
qc.x(qram[0])
qc.cswap(qram[0],ad2[1],qram[3])
qc.cswap(qram[0],ad2[2],qram[4])

qc.swap(ad2[0],qram[0])
qc.barrier()

### retieve data

qc.swap(qram[5],qram[6])
qc.swap(qram[7],qram[8])

qc.barrier()

qc.cx(qram[6],qram[4])
qc.cx(qram[8],qram[4])
qc.cx(qram[10],qram[3])
qc.cx(qram[3],qram[2])
qc.cx(qram[4],qram[2])
qc.cx(qram[2],bus[0])
qc.barrier()

# uncomputing
qc.cx(qram[3],qram[2])
qc.cx(qram[4],qram[2])
qc.cx(qram[6],qram[4])
qc.cx(qram[8],qram[4])
qc.cx(qram[10],qram[3])

#Unload address
qc.barrier()
qc.swap(qram[5],qram[6])
qc.swap(qram[7],qram[8])

qc.barrier()
qc.swap(ad2[0],qram[0])
qc.x(qram[0])
qc.cswap(qram[0],ad2[1],qram[1])
qc.cswap(qram[0],ad2[2],qram[2])
qc.x(qram[0])
qc.cswap(qram[0],ad2[1],qram[3])
qc.cswap(qram[0],ad2[2],qram[4])

qc.barrier()
qc.cx(qram[1],qram[5])
qc.cx(qram[2],qram[7])
qc.cx(qram[3],qram[9])
qc.cx(qram[4],qram[11])    

qc.barrier()

qc.x(qram[0])
qc.cswap(qram[0],ad2[1],qram[1])
qc.cswap(qram[0],ad2[2],qram[2])
qc.x(qram[0])
qc.cswap(qram[0],ad2[1],qram[3])
qc.cswap(qram[0],ad2[2],qram[4])
qc.swap(ad2[0],qram[0])
qc.x(ad2[1])
qc.cx(ad2[1],ad2[2])

qc.draw('mpl', fold=-1)
```
